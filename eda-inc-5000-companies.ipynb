 We have two types of values in revenue column
1.  revenue in Million
2.  revenue in Billion

To determine the average we must have values in either of one unit.

```python
# We are splitting the revenue field in two different fields 
companies[["revenue","Unit"]] = companies["revenue"].str.split(" ", expand= True)
```

```python
companies.head()
```

```python
companies.info()
```

Datatype of revenue field is object, to convert values representing billions in millions we need to perform some numerical operation this can only be done on numeric data

```python
# Converting data type of field
companies["revenue"] = pd.to_numeric(companies["revenue"])
```

```python
companies.info()
```

```python
#creating copy of the data and storing it in new variable.
avg_revenue = companies.copy()
```

To convert the values in billions into millions we must first determine at what index we are having entries of values in Billions.

```python
# creating a list of the index at which the revenue represents Billion value
lst = avg_revenue[avg_revenue["Unit"] == "Billion"].index
lst         
```

```python
# using for loop, multiplying revenue value at respective index with 1000
for i in lst:
    avg_revenue["revenue"].iloc[i]  = avg_revenue["revenue"].iloc[i] * 1000
```

```python
#checking whether, the operation performed is correct and as per the requirement.
avg_revenue[avg_revenue["Unit"] == "Billion"]
```

```python
# as all the values now in the revenue column is in millions, renaming this field as revenue in Millions
avg_revenue = avg_revenue.rename(columns={"revenue" : "revenue in Millions"})
avg_revenue
```

```python
# dropping out the field "Unit" as we don't need it anymore.
avg_revenue = avg_revenue.drop("Unit" , axis = 1)
```

```python
west = ["CA","CO", "HI", "ID", "MT", "NV", "OR", "UT", "WA", "WY"]
midwest = ['OH','WI', 'MI','MN','KS', 'IL','ND','MO','IN','NE', 'IA','SD']
northeast =["CT", 'DC', 'DE', 'MA','MD','ME','NH','NJ','NY','PA','RI','VT']
southeast = ['AL','AR','FL', 'GA', 'KY', 'LA', 'MS', 'NC', 'PR','SC','TN', 'VA', 'WV']
southwest = ['AZ', 'NM', 'OK', 'TX']
```

```python
def categorise(row):  
    if row['state'] in west :
        return 'west'
    elif row['state'] in midwest :
        return 'midwest'
    elif row['state'] in northeast :
        return 'northeast'
    elif row['state'] in southeast :
        return 'southeast'
    elif row['state'] in southwest :
        return 'southwest'
    
    
avg_revenue['zone'] = avg_revenue.apply(lambda row: categorise(row), axis=1)
avg_revenue
```

```python
zonewise = avg_revenue.copy()
```

```python
company_counts = zonewise.groupby(by = 'zone').size()
company_counts
```

## Around 1325 companies are in southeast region/zone which is the highest among all the regions/zones.

```python
zonewise_total_revenue = zonewise.groupby(by = 'zone').sum()
zonewise_total_revenue[["revenue in Millions"]]
```

## Companies in southeast region generates highest total revenue among all the region which is around 60267.7 Millions, followed by midwest region and then west region by generating around 59351 and 56809.4 Millions respectively.
